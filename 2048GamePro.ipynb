```python
import numpy as np

import random 
import copy
import sys
class Game:
    def __init__(self, board_size:int = 4) -> None: 
        assert isinstance(board_size, int) 
        assert board_size > 0, "0 ve 0 dan küçük olamaz"
        self.__board_size = board_size
        self.__score = 0
        self._board = [[None]*board_size for _ in range(board_size)]
        self._random_fill_cell(2)
        self._botindex=0
    
    def __emptycell(self) -> list:  
        row = []
        col=[]
        for row_index in range(self.__board_size):
            for column_index in range(self.__board_size):
                if self._board[row_index][column_index] == None:
                    row.append(row_index)
                    col.append(column_index)
        return row,col
    
    
    def _random_fill_cell(self, count:int = 1) -> None:
        assert isinstance(count,int)
        for _ in range(count):
            random_value = random.choice([2,4])
            row,column = self.__emptycell()
            r=random.randint(0,len(row)-1)
            self._board[row[r]][column[r]] = random_value
            
    def slide_r(self) -> bool:
        change = False
        
        for row_index in range(self.__board_size):
            row = [x for x in self._board[row_index][::-1] if x != None]
            for column_index in range(len(row)):
                if column_index + 1 < len(row):
                    n_index = column_index + 1
                    if row[column_index] == row[n_index]:
                        row[n_index] *= 2
                        self.__score += row[n_index]
                        row.pop(column_index)
                        change = True
            while len(row) < self.__board_size:
                row.append(None)
            if not(np.array_equiv(self._board[row_index], row[::-1])):
                change=True
          
            self._board[row_index] = row[::-1]
        
        return change
    
            
    def slide_l(self) -> bool:
        change = False
      
        for row_index in range(self.__board_size): #default:4
            row = [x for x in self._board[row_index] if x != None]
            for column_index in range(len(row)):
                if column_index + 1 < len(row):
                    n_index = column_index + 1
                    if row[column_index] == row[n_index]:
                        row[n_index] *= 2
                        self.__score += row[n_index]
                        row.pop(column_index)
                        change = True

            while len(row) < self.__board_size:
                row.append(None)
            if not(np.array_equiv(self._board[row_index], row)):
                change=True
            self._board[row_index] = row
            
                
        return change
    
    def gamestatus(self):
        control=list()
        row,column = self.__emptycell()
        if(len(row)==0):
            for row_index in range(self.__board_size):
                rowl = [x for x in self._board[row_index] if x != None]
                rowr = [x for x in self._board[row_index][::-1] if x != None]

                for lcolumn_index in range(len(rowl)):
                    if lcolumn_index + 1 < len(rowl):
                        n_index = lcolumn_index + 1
                        if rowl[lcolumn_index] == rowl[n_index]:
                            control.append("l")

                for rcolumn_index in range(len(rowr)):
                    if rcolumn_index + 1 < len(rowr):
                        n_index = rcolumn_index + 1
                        if rowr[rcolumn_index] == rowr[n_index]:
                            control.append("r")
                            
                self._board=np.transpose(self._board)
                rowl = [x for x in self._board[row_index] if x != None]
                rowr = [x for x in self._board[row_index][::-1] if x != None]
                for ucolumn_index in range(len(rowl)):
                    if ucolumn_index + 1 < len(rowl):
                        n_index = ucolumn_index + 1
                        if rowl[ucolumn_index] == rowl[n_index]:
                            control.append("u")

                for dcolumn_index in range(len(rowr)):
                    if dcolumn_index + 1 < len(rowr):
                        n_index = dcolumn_index + 1
                        if rowr[dcolumn_index] == rowr[n_index]:
                            control.append("d")
                            
                self._board=np.transpose(self._board)
            #print("Uniques= ",np.unique(control))
            if (len(np.unique(control))==0):
                print("Game Over")
                print("Your Score : ", self.__score)
                sys.exit(1)            
    
    def slide_u(self) -> bool:
        self._board=np.transpose(self._board)
        change=self.slide_l()
        self._board=np.transpose(self._board)
        return change
        
    def slide_d(self) -> bool:
        self._board=np.transpose(self._board)
        change=self.slide_r()
        self._board=np.transpose(self._board)
        return change
        
    def __str__(self) -> str:
        txt = ""
        for row_index in range(self.__board_size):
            for column_index in range(self.__board_size):
                value = str(self._board[row_index][column_index])
                value = value if value != "None" else '.'
                txt += "{:<5}".format(value)
            txt += "\n"
        txt += "\n"
        txt += f"Score : {self.__score} \n"
        return txt
    
    def playbot(self):
        self._botindex+=1
        if self._botindex%4==0:
            print("Selected Key: D\n")
            return "D"
        elif self._botindex%4==1:
            print("Selected Key: A\n")
            return "A"
        elif self._botindex%4==2:
            print("Selected Key: S\n")
            return "S"
        elif self._botindex%4==3:
            print("Selected Key: W\n")
            return "W"
        
        
# In[39]:


```
