1. Read both datasets and concatenate test set to train set. Print lenght of new dataset.

````python
import pandas as pd
import numpy as np

train = pd.read_csv("train.csv", index_col=0)
test = pd.read_csv("test.csv", index_col=0)

merge=pd.merge(train, test, left_on= ['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'], right_on= ['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],how="outer")

print(len(merge))
merge
````
