3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
df = pd.read_csv('data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
tr=df[(df['location']) == 'Turkey']

plt.figure(figsize=(20, 10))

plt.subplot(2,2,1)
sns.distplot(tr.new_deaths.dropna(), color = '#22B573', bins =50)
plt.title("kde = True", fontdict = title_style)

plt.subplot(2,2,2)
sns.distplot(tr.new_deaths.dropna(), kde = False, color = '#22B573',bins =60)
plt.title("kde = False", fontdict = title_style )

plt.subplot(2,2,3)
sns.distplot(tr.new_deaths.dropna(), color = '#22B573', bins =70)
plt.title("kde = True", fontdict = title_style)

plt.subplot(2,2,4)
sns.distplot(tr.new_deaths.dropna(), kde = False, color = '#22B573',bins =80)
plt.title("kde = False", fontdict = title_style)

plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python


locations= pd.DataFrame({'location':['Turkey', 'Germany']})
merge=pd.merge(yearfilter, locations, left_on= ['location'], right_on= ['location'],how="inner")

plt.figure(figsize=(12, 4))

plt.subplot(131)
sns.boxplot(x = 'new_deaths',  y = 'location', data = merge, palette = ['#FF914D', '#22B573'])
plt.title("x = 'new_deaths'")

plt.subplot(132)
sns.violinplot(x = 'new_deaths',  y = 'location', data = merge, palette = ['#22B573', '#FF914D'])

plt.subplot(133)
sns.swarmplot(x="location", y="new_deaths", data=merge, hue = 'location',)


plt.show()
```
