In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd
import plotly.express as px


import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

1. Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python


date=df[pd.to_datetime(df['date']) > '2020-04-01']


data = date[date.location.isin(['Spain', 'France', 'Germany', 'Italy'])]
fig = px.line(data, x="date", y="total_deaths", title='Life expectancy', color='location')
fig.show()
```

2. Plot a bar plot with plotly express for average death number that compares continents.

```python
df2=df[df.iso_code.notnull()] 
df2=df[df.continent.notnull()] 

fig = px.line(df2, x="date", y="total_deaths", title='Life expectancy', color='continent')
fig.show()
```

3. Plot a histogram for daily deaths for any country you choose. 

```python
data = df[df.location == 'Turkey']
fig = px.line(data, x="date", y="new_deaths", title='Life expectancy in Turkey')
fig.show()
```
