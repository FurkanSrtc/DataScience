1. By using distribution=np.random.name_distribution([],[]), write the name of distribution of your choice in place of `name_distribution`and fill out the bracket with your choice again. Then please, 

a) Draw the histogram and interpret

```python

import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt

data = [12,13,19,81,88,90,91,92,146,149,1,3,5,6,
                       11,44,65,79,80,100,112,113,114,120,132,
                       150,180,190,195,55,22,32,35,21,36,45,55]

result = np.random.normal(np.mean(data),np.std(data), len(data))

data.sort()
result.sort()

plt.figure(figsize=(8,4), dpi = 100)
plt.hist(result, bins=20, color = 'c')
plt.show()
```

   b) Draw Q-Q plot and interpret

```python
plt.figure(figsize=(8,5), dpi = 200)
titlefont = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
textfont  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.title('QQ', fontdict = titlefont)
plt.xlabel('data', fontdict = textfont)
plt.ylabel('result', fontdict = textfont)
plt.scatter(result, data)
plt.plot([np.min(result), np.max(data)],
         [np.min(result), np.max(data)], c='red')
plt.show()
```
