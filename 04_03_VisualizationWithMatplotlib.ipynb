In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt

df = pd.read_csv('data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)

today = pd.to_datetime('today')
daterange=pd.Timedelta('100 days')
year=pd.to_datetime(today-daterange)

df=df[pd.to_datetime(df['date']) < year]
locations= pd.DataFrame({'location':['Spain', 'France', 'Germany', 'Italy']})

merge=pd.merge(df, locations, left_on= ['location'], right_on= ['location'],how="inner")
merge

title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style2  = {'family': 'Century Gothic', 'color': 'black', 'size': 30 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }

plt.figure(figsize=(10, 7))

plt.subplot(1, 3, 1)
plt.suptitle('owid-covid-data', fontsize=30)
plt.bar(merge['location'],merge['total_deaths'], color = "#E44F28", label="Deaths" )
plt.ylabel('Total Deaths', fontdict = axis_style)
plt.title('Corona Table', fontdict = title_style)
plt.xlabel('Countries',fontdict = axis_style)
plt.legend(loc = 4)


plt.subplot(1, 3, 3)
plt.bar(merge['location'],merge['total_cases'], color = 'brown', label="Cases")
plt.ylabel('Total Cases', fontdict = axis_style2)
plt.title('Corona Table', fontdict = title_style)
plt.xticks(rotation = 90, fontsize = 20)
plt.xlabel('Countries',fontdict = axis_style)

plt.legend(loc = 4)
plt.show()

```
